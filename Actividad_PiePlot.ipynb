```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
datos1 = pd.read_csv("C:\\FCD\\PythonFundamental_Femsa\\athlete_events.csv")
datos1
```

```python
#Crear un programa en Visual Studio que me permita saber cuÃ¡l es el competidor mÃ¡s veterano que ha recibido medalla (oro, plata o bronce)
anomax = datos1[datos1["Year"].notna()].max() #verifica o maior ano da base, nesse exemplo: 2016
anomax
```

```python
#Filtra dados de 2016
ano16 = datos1[datos1["Year"]==2016]
ano16
```

```python

#Filtra medalhas de 2016
Gold16=ano16[ano16["Medal"] == "Gold"] #filtra onde a medalha seja Ouro
Silv16=ano16[ano16["Medal"] == "Silver"] #filtra onde a medalha seja Ouro
Bron16=ano16[ano16["Medal"] == "Bronze"] #filtra onde a medalha seja Ouro
GS16 = Gold16.append(Silv16)
GSB16 = GS16.append(Bron16) #unifica as bases de medalha de 2016
```

```python
#agrupa medalhas por paÃ­s (NOC)
noc_med=GSB16.groupby("NOC").count()
noc_med["Medal"]
```

```python
#Cria grÃ¡fico de pizza (Pie Plot)
pie = noc_med["Medal"].plot.pie(figsize=(12,12))
```
